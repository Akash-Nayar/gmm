```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
import math
# np.set_printoptions(threshold=np.inf)
```

```python
class GMM():
    def __init__(self):
        self.is_fit = False
        
    
    def calculate_weights(self, X, pi, mu, sigma, normalize=True):
        assert(len(pi) == len(mu) == len(sigma))
        centered = np.array([X - mu[i] for i in range(len(mu))])
        first = lambda i: np.dot(centered[i], np.linalg.inv(sigma[i]))[:, None, :]
        second = lambda i: np.matrix.transpose(centered[i][:, None], (0, 2, 1))
        w = np.exp(np.squeeze(-0.5 * np.array([first(i) @ second(i) for i in range(len(mu))])))
        return w / np.sum(w, axis=0)[None, :] if normalize else w
    
    
    def predict(self, X):
        assert(self.is_fit and len(self.mu[0]) == X.shape[1])
        return np.argmax(self.calculate_weights(X, self.pi, self.mu, self.sigma), axis=0)
        
    
    def fit(self, X, n_clusters, max_retries=10, epsilon=0.001):
        assert(n_clusters > 0)
        self.n_clusters = n_clusters
        b_likelihood = float("inf")
        for f in range(max_retries):
            np.random.shuffle(X)
            pi = np.ones((n_clusters, 1)) / n_clusters
            mu = X[:n_clusters]
            sigma = np.array([np.cov(sub.T) for sub in np.array_split(X, n_clusters)])
            raw_w = None
            for fg in range(20): # replace with termination when loss < epsilon
                raw_w = self.calculate_weights(X, pi, mu, sigma, normalize=False)
                w = raw_w / np.sum(raw_w, axis=0)[None, :]
                n = np.sum(w, axis=1)[:, None]
                pi = n / np.sum(n)
                mu = (w @ X) / n
                centered = np.array([X - mu[i] for i in range(len(mu))])
                second = lambda i: centered[i][:, None, :]
                first = lambda i: np.matrix.transpose(centered[i][:, None], (0, 2, 1))
                for j in range(n_clusters):
                    sigma[j] = np.sum(first(j) @ second(j) * w[j][:, None, None], axis=0) / n[j]
#             print(np.sum(raw_w))
            if (likelihood := np.sum(raw_w)) < b_likelihood:
                b_likelihood = likelihood
                self.pi = pi
                self.mu = mu
                self.sigma = sigma
        self.is_fit = True
```

```python
np.random.seed(42)
```

```python
cluster1 = np.random.normal(size=(200, 2)) * 0.3 - np.array([4, -0.5])
cluster2 = np.random.normal(size=(100, 2)) * 0.5 - np.array([-0.5, -0.5])
cluster3 = np.random.normal(size=(50, 2)) * 0.2 - np.array([0, 5])
cluster4 = np.random.normal(size=(250, 2)) * 0.2 - np.array([5, 5])
data = np.vstack((cluster1, cluster2, cluster3, cluster4))
```

```python
plt.scatter(*(data.T))
```

```python
gmm = GMM()
gmm.fit(data, 4, max_retries=500)
```

```python
print(gmm.sigma)
print(gmm.pi)
print(gmm.mu)
```

```python
color_map = ['red', 'green', 'blue', 'purple']
colors = []
for pred in gmm.predict(data):
    colors.append(color_map[pred])
```

```python
plt.scatter(*(data.T), color=colors)
for i in range(4):
    plt.scatter(gmm.mu[:, 0], gmm.mu[:, 1], color=color_map, marker='x')
```
